## Exercise 2.2

Write an algorithm that reads three different integer values ​​and displays them in descending order. Use nested conditional statements for this purpose.

Initialization

```python
a = int(input("Number a: "))
b = int(input("Number b: "))
c = int(input("Number c: "))
```

### First version

```python
if (a > b) and (b > c):
    print(a, b, c)
if (a > c) and (c > b):
    print(a, c, b)
if (b > a) and (a > c):
    print(b, a, c) 
if (b > c) and (c > a):
    print(b, c, a)
if (c > a) and (a > b):
    print(c, a, b)
if (c > b) and (b > c):
    print(c, b, a) 
```

### Second version

Optimized version

```python
nums = [a, b, c]
nums.sort(reverse=True)
print(*nums)
```

## Exercise 2.3

Develop an algorithm that calculates the roots of a quadratic equation in the form 

$$ax^2 + bx + c = 0$$

taking into account the existence of real roots.

```python
import math

def roots(a, b, c):
    """Calculate real roots of ax^2 + bx + c = 0"""
    delta = b**2 - 4 * a * c

    if delta < 0:
        print("The roots are not real.")
        return None
    else:
        x1 = (-b + math.sqrt(delta))/2*a
        x2 = (-b - math.sqrt(delta))/2*a
        return x1, x2
```

Example 1: real roots

```python
a = 1
b = -3
c = 2
```

```python
print(f"Roots: {roots(a, b, c)}")
```

 Example 2: The roots of the equation are not real.

```python
a = 2
b = 1
c = 5
```

```python
print(f"Roots: {roots(a, b, c)}")
```

## Exercise 2.4

Given a person's height, $h$, and sex as input, create an algorithm that calculates their ideal weight using the following formulas:

- For men: $72.7h - 58$
- For women: $62.1h - 44.7$

Weight is m

```python
def ideal_weight(h, sex):
    """Calculate ideal weight"""
    if sex == "h":
        return 72.7*h - 58
    else:
        return 62.1*h - 44.7
```

Firs example:

- Men
- 1.7 m

```python
print(f"Ideal weight: {ideal_weight(1.7, 'h')} kg")
```

Second example:

- Women
- 1.65 m

```python
print(f"Ideal weight: {ideal_weight(1.65, 'm')} kg")
```
